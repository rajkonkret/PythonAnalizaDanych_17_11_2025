```python
temp_array = []
with open('data/AAPL_stock_price_example.csv', "r") as input_file:
    all_lines = input_file.readlines()
    for each_line in all_lines:
        for value_from_line in each_line.rsplit(",")[1:]:
            clean_value = value_from_line.replace("\n", "")
            temp_array.append(clean_value)
```

```python
print(temp_array[:5])
```

```python
print(len(temp_array))
```

```python
temp_array = np.delete(temp_array, 0)  # usunięcie pierwszego wiersza z nazwami kolumn
```

```python
temp_array.size
```

```python
temp_array.dtype
```
