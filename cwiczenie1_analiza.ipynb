ravel() - zwraca widok\
operacje na widoku "tablicy" wpływaja na zmainy oryginalnej tablicy

```python
print(new_arr.ravel())
```

flatten() - robi kopię\
zmiany nie wpływają na oryginlaną tablicę

```python
print(new_arr.flatten())
```

```python
flat = new_arr.flatten()  # kopia
rav = new_arr.ravel()  # widok
```

```python
flat[0] = 999  # kopia, nie zmienia oryginału
rav[1] = 888  # widok, wpływa na oryginał
```
