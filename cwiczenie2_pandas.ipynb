```python
movies_db.drop_duplicates(subset=["Director"])  # usuniecie gdy dane w kolumnie się powtarzają
```

```python
rev = movies_db['Revenue (Millions)']
rev.head()
```

```python
movies_db.columns = [col.lower() for col in movies_db]
movies_db.tail()  # 5 ostatnich
```

```python
movies_db.describe()
```

```python
# sortowanie kolumny po rating
movies_db.sort_values(by="rating")
```

```python
# count - liczy wszystkie bez NaN
movies_db.groupby(['genre']).agg({"year": "count"}).sort_values(by="year", ascending=False)
```

```python
result = movies_db.groupby(['genre']).agg({"year": "count"}).sort_values(by="year", ascending=False)
result = result.rename(columns={"year": "num_movies"})
result
```

```python
movies_db.groupby('genre').agg(num_movies=("year", "count")).sort_values(by="num_movies", ascending=False)
```

```python
store1 = pd.read_csv("data/Store1.csv")
store2 = pd.read_csv("data/Store2.csv")
```

```python
store1.head()
```

```python
store2.tail()
```

```python
sales_store1 = (store1[["Grocery_sales", "Stationary_sales"]].sum().sum())
sales_store1
```

```python
month_sales1 = store1.groupby("Months")[["Grocery_sales", "Stationary_sales"]].agg("sum").agg("sum", axis=1)
month_sales1
```

```python
month_sales1a = store1.groupby("Months")[["Grocery_sales", "Stationary_sales"]].agg("sum").agg("sum", axis=0)
month_sales1a
```
